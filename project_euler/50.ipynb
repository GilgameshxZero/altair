```python
LIMIT = 1000000
is_prime = [False] * LIMIT
primes = []
for i in range(2, LIMIT):
	is_prime[i] = True
	for j in primes:
		if j * j > i:
			break
		if i % j == 0:
			is_prime[i] = False
			break
	if is_prime[i]:
		primes.append(i)
primes_set = set(primes)
primes[:8]
```

```python
z = 21
for i in range(len(primes)):
	sum = primes[i]
	for j in range(i + 1, len(primes)):
		sum += primes[j]
		if sum > primes[-1]:
			break
		if j - i + 1 > z and sum in primes_set:
			z = j - i + 1
			print(z, sum)
```
