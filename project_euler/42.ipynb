```python
with open(utulek.ASSET_PATH +
	".data/words.txt") as file_in:
	words = [
		"".join(j if j != "\"" else "" for j in i)
		for i in file_in.read().split(",")
	]
len(words), words[400]
```

```python
tri_nums = [i * (i + 1) // 2 for i in range(0, 100)]


def is_tri_word(word):
	value = sum(ord(i) - ord("A") + 1 for i in word)
	return value in tri_nums


is_tri_word("HELLO"), is_tri_word("SKY")
```

```python
sum(is_tri_word(i) for i in words)
```
