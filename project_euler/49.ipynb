```python
LIMIT = 9999
is_prime = [False] * LIMIT
primes = []
for i in range(2, LIMIT):
	is_prime[i] = True
	for j in primes:
		if j * j > i:
			break
		if i % j == 0:
			is_prime[i] = False
			break
	if is_prime[i]:
		primes.append(i)
primes[:8]
```

```python
primes_set = set(primes)
```

```python
for i in range(1000, 9999):
	for j in range(1, 9999):
		if i + 2 * j > 9999:
			break
		nums = [i + k * j for k in range(0, 3)]
		if any([k not in primes_set for k in nums]):
			continue
		num_strs = [str(k) for k in nums]
		ns_sorted = [sorted(k) for k in num_strs]
		if all([k == ns_sorted[0] for k in ns_sorted]):
			print(i, j)
			break
```
