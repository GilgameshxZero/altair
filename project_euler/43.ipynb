```python
def is_valid(x):
	sub_nums = [
		int(x[i] + x[i + 1] + x[i + 2])
		for i in range(0,
		len(x) - 2)
	]
	divisors = [1, 2, 3, 5, 7, 11, 13, 17]
	return not any([
		sub_nums[i] % divisors[i]
		for i in range(0, len(divisors))
	])


is_valid("1406357289")
```
