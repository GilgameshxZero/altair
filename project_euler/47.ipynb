```python
LIMIT = 999
is_prime = [False] * LIMIT
primes = []
for i in range(2, LIMIT):
	is_prime[i] = True
	for j in primes:
		if j * j > i:
			break
		if i % j == 0:
			is_prime[i] = False
			break
	if is_prime[i]:
		primes.append(i)
primes[:8]
```

```python
def count_prime_factors(n):
	z = 0
	for i in primes:
		z += n % i == 0
	return z


count_prime_factors(644)
```

```python
c_prime_factors = [
	count_prime_factors(i) for i in range(0, LIMIT**2)
]
```

```python
for i in range(len(c_prime_factors)):
	if all(
		[c_prime_factors[j] == 4 for j in range(i, i + 4)]):
		print(i)
		break
```
